```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import scipy as sp
import faraway.utils
```

```python
ds = sm.datasets.get_rdataset("ethanol", "lattice")
ethanol = ds.data
ethanol.head()
```

```python
print(ds.__doc__)
```

```python
sns.scatterplot(y="NOx", x="E", data=ethanol, hue="C")
```

Plot indicates a sort of optimal value around 0.9 </br>
Therefore, a reasonable choice -- fit a polynomial in E

```python
import faraway.utils
polmod = smf.ols("NOx ~ E + I(E**2) + C", data=ethanol).fit()
polmod.sumary()
```

```python
f, ax = plt.subplots(1, 1)
sns.scatterplot(y="NOx", x="E", data=ethanol, ax=ax)
xlim = ax.get_xlim()
xr = np.linspace(xlim[0], xlim[1], num=50)
ax.plot(
    xr, 
    polmod.params[0] + 
    xr * polmod.params[1] + 
    xr**2 * polmod.params[2] + 
    ethanol["C"].mean()*polmod.params[3]
)
```

```python
-polmod.params[1]/(2*polmod.params[2])
```

```python
f2 = np.poly1d(polmod.params[[2, 1, 0]])
from scipy.optimize import minimize_scalar
result = minimize_scalar(-f2)
result.x
```

```python
sns.scatterplot(y=polmod.resid, x=ethanol["E"])
```

The plot indicates some remaining structure that suggests a more flexible
function than the quadratic is required.

```python
polmod2 = smf.ols("NOx ~ E + I(E**2) + I(E**3) + I(E**4) + C", data=ethanol).fit()
polmod2.sumary()
```

```python
sns.scatterplot(y=polmod2.resid, x=ethanol["E"])
plt.axhline(0, ls='--')
```

The plot is now ok

In order to ensure that the response is always predicted positive, we can log it.

```python
polmod3 = smf.ols("np.log(NOx) ~ E + I(E**2) + C", data=ethanol).fit()
polmod3.sumary()
```

The fit is better then a regular parabola

```python
sns.scatterplot(y=polmod3.resid, x=ethanol["E"])
plt.axhline(0, ls='--')
```

Additional structure is still present though.

```python
f, ax = plt.subplots(1, 1)
sns.scatterplot(y="NOx", x="E", data=ethanol, ax=ax)
xlim = ax.get_xlim()
xr = np.linspace(xlim[0], xlim[1], num=50)

ax.plot(
    xr, 
    polmod.params[0] + 
    xr * polmod.params[1] + 
    xr**2 * polmod.params[2] + 
    ethanol["C"].mean()*polmod.params[3],
    label="quadratic poly"
)

ax.plot(
    xr, 
    polmod2.params[0] + 
    xr * polmod2.params[1] + 
    xr**2 * polmod2.params[2] + 
    xr**3 * polmod2.params[3] + 
    xr**4 * polmod2.params[4] + 
    ethanol["C"].mean()*polmod2.params[5],
    ls='--',
    label="quartic poly"
)

ax.plot(
    xr, 
    np.exp(
        polmod3.params[0] + 
        xr * polmod3.params[1] + 
        xr**2 * polmod3.params[2] + 
        ethanol["C"].mean()*polmod3.params[3]
    ),
    ls="dotted",
    label="logged poly"
)
plt.legend()
```
