## Error analysis ##

```python
resid_yhat_plot(lmod)
```

```python
ddf = pd.DataFrame({
    "x": lmod.fittedvalues, 
    "y": np.sqrt(np.abs(lmod.resid))
})
smf.ols(
    "y ~ x",
    data=ddf
).fit().sumary()
```

Normality and homoscedacity are violated. Possible solution: transformation of the response.

```python
lmod_tr = smf.ols(
    "np.sqrt(gamble) ~ sex + status + income + verbal",
    data=teengamb
).fit()
lmod_tr.sumary()
```

```python
left_group = lmod.resid[lmod.fittedvalues < 10]
right_group = lmod.resid[lmod.fittedvalues > 10]

_ = resid_yhat_plot(lmod_tr, to_annotate=left_group.index)
```

```python
lmod_tr.resid.argmax()
```

```python

# f-test
fstat = np.var(right_group, ddof=1)/ np.var(left_group, ddof=1)
2 * (1-sp.stats.f.cdf(
    fstat, 
    len(right_group)-1,
    len(left_group)-1) 
)
```

```python
# or, equiv.
fstat = np.var(left_group, ddof=1) / np.var(right_group, ddof=1)
2 * sp.stats.f.cdf(
    fstat, 
    len(left_group)-1,
    len(right_group)-1
)
```

The difference was significant

```python
left_group_tr = lmod_tr.resid[lmod.fittedvalues < 2.5]
right_group_tr = lmod_tr.resid[lmod.fittedvalues > 2.5]

fstat = np.var(left_group_tr, ddof=1) / np.var(right_group_tr, ddof=1)
2 * sp.stats.f.cdf(
    fstat, 
    len(left_group_tr)-1,
    len(right_group_tr)-1
)
```

Not significant anymore.

### Model selection

```python
lmod.sumary()
X = pd.DataFrame(lmod.model.exog, columns=lmod.params.index)
y = lmod.model.endog
```

```python
import itertools
pcols = list(X.columns.drop(["sex", "Intercept"]))

pd.DataFrame(
    np.empty((len(pcols)+1), 4),
    columns=["rss", "pval", "aic", "r2adj"]
)
rss = np.empty(len(pcols) + 1)
pval = np.empty(len(pcols) + 1)

selvar = [('sex', 'Intercept')]
rss[0] = sm.OLS(y, X.loc[:, selvar[0]]).fit().ssr
pval[0] = sm.OLS(y, X.loc[:, selvar[0]]).fit().pvalues["sex"]

for k in range(1, len(pcols)+1):
    RSS = {}
    PVAL = {}
    for variables in itertools.combinations(pcols, k):
        variables = list(variables)
        variables.extend(["sex", "Intercept"])
        predictors = X.loc[:, variables]
        res = sm.OLS(y, predictors).fit()
        RSS[tuple(variables)] = res.ssr
        PVAL[tuple(variables)] = res.pvalues["sex"]
    rss[k] = min(RSS.values())
    pval[k] = min(PVAL.values())
    selvar.append(min(RSS, key=RSS.get))
    
min_rss_models = list(zip(selvar, rss.round(3), pval.round(3)))
min_rss_models
```

```python
import itertools
pcols = list(X.columns.drop(["sex", "Intercept"]))

res = pd.DataFrame(columns=["rss", "pval"])

for k in range(0, len(pcols)+1):
    for variables in itertools.combinations(pcols, k):
        variables = list(variables)
        variables.extend(["sex", "Intercept"])
        predictors = X.loc[:, variables]
        
        lmodi = sm.OLS(y, predictors).fit()
        resi = pd.DataFrame({
                "#pred": k+1,
                "rss": lmodi.ssr, 
                "pval": lmodi.pvalues["sex"],
                "aic": lmodi.aic,
                "r2adj": lmodi.rsquared_adj,
            }, 
            index=[str(variables)]
        )
        res = pd.concat([res, resi])
res
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.lineplot(data=res.groupby("#pred")["rss"].min(), ci=None, ax=ax1)
sns.lineplot(data=res.groupby("#pred")["r2adj"].max(), ci=None, ax=ax2)

for ax in [ax1, ax2]:
    ax.set_xticks(range(1, int(res["#pred"].max())+1))

print(
    "Suggested by AIC:", res.iloc[res["aic"].argmin(), :].name, '\n',
    "Suggested by R^2adj:", res.iloc[res["r2adj"].argmax(), :].name
)
```

Models with `income` are distinctively better.
